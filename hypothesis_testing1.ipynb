
##### 1. Assume they follow a normal distribution and we assume equal variance, we will like to know if there is a difference between both sample means. We will perform a two-sample t-test for independent samples



```python
sample1 = np.array([9, 11, 10,11,10,12, 9,11,12, 9, 10])
sample2 = np.array([10, 13, 10, 13, 12, 9, 11, 12, 12, 12, 13])
```

To test whether the variances of two samples are the same, we use Levene's test.

```python
scipy.stats.levene(sample1,sample2,center = 'mean')
```

p_value from the Levene's test are larger than 0.05, so we will assume that variances are equal between 2 samples.

```python
scipy.stats.ttest_ind(sample1,sample2,equal_var = True)
```

p_value of simple t test is less than 0.05, so we can reject the null hypothesis,there is a difference in sample means between 2 samples.

#####2. A room  in a laboratory is only considered safe if the mean radiation level is 400 or less. When a sample of 10 radiation measurements were taken, the mean value of the radiation was 414 with a standard deviation of 17. There are concerns that mean radiation is above 414. Radiation levels in the lab are known to follow a normal distribution with standard deviation 22. We will like to conduct a hypothesis test at the 5% level of signiÔ¨Åcance to determine whether there is evidence that the laboratory is unsafe.

Because the standard deviation of the population is known, we are going to use Z-test for this problem. 

Null hypothesis: there is  no difference between mean of radiation level of the sample compared to that of the population
Alternative hypothesis: mean radiation level of the sample is higher than that of the population.



```python
scipy.stats.norm.sf((414-400)/(22/sqrt(10)))
```

```python
1 - scipy.stats.norm.cdf((414-400)/(22/sqrt(10)))
```

###### Methods

```python
# Caculate the p_value knowing the Z-score
print(scipy.stats.norm.sf(abs(1.96))) #one-sided
# Calculate the probability (cumulative area) using Z score
print(scipy.stats.norm.cdf(1.64))
# Caculating the critical value knowing probability
print(scipy.stats.norm.ppf(.95))
```
