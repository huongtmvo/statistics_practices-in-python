```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
# Create data with normal distribution
r = 1
mean = [15, 5]
cov = [[1, r], [r, 1]]
x, y = x, y = np.random.multivariate_normal(mean, cov, 400).T
```

```python
plt.figure(figsize=(15,10))
```

```python
plt.subplot(1,2,1)
plt.hist(x = x, bins = 20)
plt.title("X distribution")

plt.subplot(1,2,2)
plt.hist(x = y, bins = 20)
plt.title(" Y distribution")

plt.show()
```

```python
plt.figure(figsize = (15,5))

plt.subplot(2,2,1)
plt.scatter(x = x, y = y)
plt.title(' Joint distribution of X and Y')

plt.subplot(2,2,2)
plt.hist(x = x, bins = 20)
plt.title ('Marginal distribution of X')

plt.subplot(2,2,3)
plt.hist(x = x, orientation = 'horizontal', bins = 20)
plt.title('Marginal distribution of Y')
plt.show()
```
