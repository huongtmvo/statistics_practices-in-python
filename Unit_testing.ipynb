## Unit Testing
While we will not cover the [unit testing library](https://docs.python.org/3/library/unittest.html) that python has, we wanted to introduce you to a simple way that you can test your code.

Unit testing is important because it the only way you can be sure that your code is do what you think it is doing. 

Remember, just because ther are no errors does not mean your code is correct.

### Goal
We want to find the mean of first 100 rows of 'BPXSY1' when 'RIDAGEYR' > 60

```python
# One possible way of doing this is:
pd.Series.mean(df[df.RIDAGEYR > 60].loc[0:100, 'BPXSY1']) 
 # or 
df[df.RIDAGEYR > 60].loc[0:100,'BPXSY1'].mean()

# These ways are WRONG, we will examinate them further
# The size of the data frame above will not be 100
df[df.RIDAGEYR > 60].loc[0:100,'BPXSY1'].shape 
# There are only 28 rows that satisfy the condition, therefore will give incorrect results. 
```

```python
df[df.RIDAGEYR > 60].iloc[range(0,5), :] # Correct picks the first five rows such that 'RIDAGEYR" > 60
```

```python
# This correctly picked the 1st 100 rows with RIDAGEYR > 60
df[df.RIDAGEYR > 60].iloc[range(0,100), 16].shape
```

```python
# First way: applying the correct method to the original question about BPXSY1
print(pd.Series.mean(df[df.RIDAGEYR > 60].iloc[range(0,100), 16]))

# Second way: reference the BPXSY1 variable
print(pd.Series.mean(df[df.RIDAGEYR > 60].iloc[range(0,100), df.columns.get_loc('BPXSY1')]))

# Third way:
print(df[df.RIDAGEYR > 60].iloc[range(0,100),16].mean())

# Fourth way:  
print(df[df.RIDAGEYR > 60].loc[:,'BPXSY1'].head(100).mean())

# df.columns.get_loc('BPXSY1') to find the index of column BPXSY1
```
